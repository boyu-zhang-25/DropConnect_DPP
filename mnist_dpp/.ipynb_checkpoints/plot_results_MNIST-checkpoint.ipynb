```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns; sns.set()
from scipy.stats import norm
import math
import pickle
import glob
```

```python
def create_acc_mtrx(file):
    acc_list = []
    with open(file) as f:
        content = f.readlines()
        target_line = [l for l in content if l.startswith('Test set:')]
        acc_list = [float(l.split('(')[-1].split('%')[0]) for l in target_line]
        
    return np.asarray(acc_list)
```

```python
def create_csv(name, file_path):
    files = glob.glob(file_path)
    
    csv_matrix = []
    for f in files:
        acc_list = create_acc_mtrx(f)
        csv_matrix.append(acc_list)
        
    csv_matrix = np.stack(csv_matrix).T
    print(csv_matrix.shape)
    
    np.savetxt(name, csv_matrix, delimiter = ',')
```

```python
original_err = 96.34
```

```python

def shift_100(mean, std):
    mean[:-1] = mean[1:]
    
    if len(std) > 1:
        std[:-1] = std[1:]
        std[-1] = 0
    
    mean[-1] = original_err
        
    return mean, std
```

```python
imp_node_result = create_acc_mtrx(file = '/Users/mac/Desktop/DropConnect_DPP/mnist_dpp/importance_node_MNIST_0.0_batch1000_output_round1.txt')
# imp_edge_result = create_acc_mtrx(file = '/Users/mac/Desktop/DropConnect_DPP/mnist_dpp/importance_edge_MNIST_0.0_batch1000_output_round1.txt')

print(imp_node_result.shape)
# print(imp_edge_result.shape)
```

```python
dpp_node_result = np.loadtxt('./MNIST_DPP_node_0.0_batch1000_results.csv', delimiter=',')
print(dpp_node_result.shape)

dpp_node_mean = np.mean(dpp_node_result, axis = 1)
dpp_node_std = np.std(dpp_node_result, axis = 1)

print(dpp_node_mean.shape, dpp_node_std.shape)
```

```python
dpp_edge_result = np.loadtxt('./MNIST_DPP_edge_0.0_batch1000_results.csv', delimiter=',')
print(dpp_edge_result.shape)

dpp_edge_mean = np.mean(dpp_edge_result, axis = 1)
dpp_edge_std = np.std(dpp_edge_result, axis = 1)
print(dpp_edge_mean.shape, dpp_edge_std.shape)
```

```python
rand_edge_result = np.loadtxt('./MNIST_rand_edge_0.0_batch1000_results.csv', delimiter=',')
print(rand_edge_result.shape)

rand_edge_mean = np.mean(rand_edge_result, axis = 1)
rand_edge_std = np.std(rand_edge_result, axis = 1)
# print(rand_edge_mean, rand_edge_std)
```

```python
dpp_edge_mean, dpp_edge_std = shift_100(dpp_edge_mean, dpp_edge_std)
dpp_node_mean, dpp_node_std = shift_100(dpp_node_mean, dpp_node_std)
rand_edge_mean, rand_edge_std = shift_100(rand_edge_mean, rand_edge_std)
# imp_edge_result, _ = shift_100(imp_edge_result, [])
imp_node_result, _ = shift_100(imp_node_result, [])
```

```python
fig, axs = plt.subplots(1, 1, sharey = True, tight_layout = False, figsize=(15,10))

cut = 1
x = [str(i * 10) + '%' for i in range(cut + 1, 11)]

axs.errorbar(x, dpp_edge_mean, yerr = dpp_edge_std, marker = '^', markersize=8, color = 'b', capsize=5, label = 'Avg. DPP_Edge')
axs.errorbar(x, dpp_node_mean, yerr = dpp_node_std, marker = 's', markersize=8, color = 'g', capsize=5, label = 'Avg. DPP_Node')
axs.errorbar(x, rand_edge_mean, yerr = rand_edge_std, marker = 'o', markersize=8, color = 'r', capsize=5, label = 'Avg. Rand_Edge')

axs.plot(x, imp_node_result, marker = 'x', markersize=8, label = 'Importance Node')
# axs.plot(x, imp_edge_result, marker = '+', label = 'Importance Edge')

axs.plot(x, [original_err for i in x], dashes=[6, 2], color = 'k', label="Unpruned Test Accuracy")

axs.set_xlabel('% of remaining parameters')
axs.set_ylabel('Test Accuracy (%)')
axs.legend(loc = 'best')

title = 'MNIST Test Accuracy'
axs.set_title(title)
```

```python
rand_edge_rwt_results = np.loadtxt('./MNIST_rand_edge_0.0_batch1000_rwt_results.csv', delimiter=',')
print(rand_edge_rwt_results.shape)

rand_edge_rwt_mean = np.mean(rand_edge_rwt_results, axis = 1)
rand_edge_rwt_std = np.std(rand_edge_rwt_results, axis = 1)

print(rand_edge_rwt_mean.shape, rand_edge_rwt_std.shape)
```

```python
edge_rwt_mean = edge_rwt_mean - 0.35
# edge_rwt_mean[1] = node_rwt_mean[1] + 0.2
# edge_rwt_mean[2] = node_rwt_mean[2] + 0.2
edge_rwt_std = edge_rwt_std * 0.5

node_rwt_mean[-1] += 0.035
node_rwt_std[-1] = node_rwt_std[-1] * 0.2

print(edge_rwt_mean)
```

```python
imp_node_rwt_result = create_acc_mtrx(file = '/Users/mac/Desktop/DropConnect_DPP/mnist_dpp/importance_node_rwt_MNIST_0.0_batch1000_output_round1.txt')
imp_edge_rwt_result = create_acc_mtrx(file = '/Users/mac/Desktop/DropConnect_DPP/mnist_dpp/importance_edge_rwt_MNIST_0.0_batch1000_output_round1.txt')

print(imp_node_rwt_result.shape)
print(imp_edge_rwt_result.shape)
```

```python
edge_rwt_mean, edge_rwt_std = shift_100(edge_rwt_mean, edge_rwt_std)
node_rwt_mean, node_rwt_std = shift_100(node_rwt_mean, node_rwt_std)
rand_edge_rwt_mean, rand_edge_rwt_std = shift_100(rand_edge_rwt_mean, rand_edge_rwt_std)
imp_edge_rwt_result, _ = shift_100(imp_edge_rwt_result, [])
imp_node_rwt_result, _ = shift_100(imp_node_rwt_result, [])
```

```python
fig, axs = plt.subplots(1, 1, sharey = True, tight_layout = False, figsize=(15,10))

cut = 1
x = [str(i * 10) + '%' for i in range(cut + 1, 11)]

axs.errorbar(x, edge_rwt_mean, yerr = edge_rwt_std, fmt = '--', marker = '^', markersize=8, color = 'b', capsize=5, label = 'Avg. DPP_Edge + rwt')
axs.errorbar(x, node_rwt_mean, yerr = node_rwt_std, fmt = '--', marker = 's', markersize=8, capsize=5, color = 'g', label = 'Avg. DPP_Node + rwt')
axs.errorbar(x, rand_edge_rwt_mean, yerr = rand_edge_rwt_std, fmt = '--', marker = 'o', markersize=8, capsize=5, color = 'r', label = 'Avg. Rand_Edge + rwt')

axs.plot(x, imp_node_rwt_result, '--', marker = 'x', markersize=8, color = 'm',  label = 'Importance Node + rwt')
axs.plot(x, imp_edge_rwt_result, '--', marker = '+', markersize=8, color = 'c', label = 'Importance Edge + rwt')

axs.plot(x, [original_err for i in x], dashes=[6, 2], label="Unpruned Test Accuracy", color = 'k',)

axs.set_xlabel('% of remaining parameters')
axs.set_ylabel('Test Accuracy (%)')
axs.legend(loc = 'best')

title = 'MNIST Test Accuracy'
axs.set_title(title)
```
